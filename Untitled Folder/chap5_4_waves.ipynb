## First example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 24)
t
```

```python
A = 1.25 # [m] the amplitude
T = 0.5 # [d] cycle time
kD = 600 # [m2/d], transmissivty
S = 0.1 # [-], storage coefficient
omega = 2 * np.pi / T  # [radians/d] angle velocity
a = np.sqrt(omega * S  / (2 * kD)) # [1/m] damping factor
x = np.linspace(0, 500, 501)
times = np.arange(0, 24) / 24

plt.title('My first example\nin class\ncourse 2019-20120')
plt.xlabel('distance [d]')
plt.ylabel('drawdown s(x, t) [m]')
plt.plot(x, +A * np.exp(-a * x), label='Upper envelope')
plt.plot(x, -A * np.exp(-a * x), label='Lower envelope')

for t in times:
    y = A * np.exp(-a * x) * np.sin(omega * t - a * x)
    plt.plot(x, y, label='t = {:4g}'.format(t))

plt.legend()
plt.grid()
print("we're done")
```

```python
As = np.array([1.25, 0.6, 1.7]) # [m] the amplitude
T = np.array([0.5 / 24, 0.5, 5]) # [d] cycle time
kD = 600 # [m2/d], transmissivty
S = 0.1 # [-], storage coefficient
omegas = 2 * np.pi / T  # [radians/d] angle velocity
a3 = np.sqrt(omegas * S  / (2 * kD)) # [1/m] damping factor
x = np.linspace(0, 500, 501)
times = np.arange(0, 24, 6) / 24

plt.title('My first example\nin class\ncourse 2019-20120')
plt.xlabel('distance [m]')
plt.ylabel('drawdown s(x, t) [m]')
#plt.plot(x, +A * np.exp(-a * x), label='Upper envelope')
#plt.plot(x, -A * np.exp(-a * x), label='Lower envelope')


t = times[3]
Y = np.zeros_like(x)
for omega, A, a in zip(omegas, As, a3):
    y = A * np.exp(-a * x) * np.sin(omega * t - a * x)
    Y = Y + y
    plt.plot(x, y, label='t = {:4g}'.format(t))

plt.plot(x, Y, lw=3, label='total', alpha=0.3)
plt.legend()
plt.grid()
print("we're done")
```
